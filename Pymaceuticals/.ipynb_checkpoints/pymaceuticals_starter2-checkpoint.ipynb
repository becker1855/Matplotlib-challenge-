```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df= pd.read_csv(mouse_drug_data_to_load)
clinical_trial_df= pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
merged_data_df = pd.merge(mouse_drug_df, clinical_trial_df, on="Mouse ID")

# Display the data table for preview
merged_data_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
merged_first= merged_data_df.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"]
avg_tumor= merged_first.mean()
# Convert to DataFrame
tumor_response= avg_tumor.reset_index()
# Preview DataFrame
tumor_response.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
merged_second= merged_first.sem()
# Convert to DataFrame
tumor_stderror= pd.DataFrame(merged_second)
# Preview DataFrame
tumor_stderror.reset_index(inplace=True)
tumor_stderror.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivot_merged= tumor_response.pivot(index= "Timepoint", columns="Drug", values = "Tumor Volume (mm3)")
pivot_merged.reset_index(level= None, inplace= True)
# Preview that Reformatting worked
pivot_merged
```

```python
reduced_drugs = pivot_merged[["Timepoint", "Capomulin", "Infubinol", "Ketapril", "Placebo"]]
reduced_drugs
```

```python
# Generate the Plot (with Error Bars)
plt.figure(figsize= (10,5))
plt.errorbar(x=reduced_drugs['Timepoint'],y=reduced_drugs['Capomulin'], yerr=None, linestyle="-", fmt='o', label="Capomulin")
plt.errorbar(x=reduced_drugs['Timepoint'],y=reduced_drugs['Infubinol'], yerr=None, linestyle="-", fmt='o', label="Infubinol")
plt.errorbar(x=reduced_drugs['Timepoint'],y=reduced_drugs['Ketapril'], yerr=None, linestyle="-",fmt='o', label="Ketapril")
plt.errorbar(x=reduced_drugs['Timepoint'],y=reduced_drugs['Placebo'], yerr=None, linestyle="-", fmt='o', label="Placebo")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Time(Days)")
plt.title("Tumor Response to Treatment")
plt.legend()
# Save the Figure
plt.save(image/figure1)
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
merge_third= merged_data_df.groupby(["Drug", "Timepoint"])["Metastatic Sites"]
avg_meta= merge_third.mean()
# Convert to DataFrame
metastatic_response= avg_meta.reset_index()
# Preview DataFrame
metastatic_response.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
merge_fourth= merge_third.sem()
# Convert to DataFrame
meta_stderror= pd.DataFrame(merge_fourth)
# Preview DataFrame
meta_stderror.reset_index()
meta_stderror.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivot_merged_two= metastatic_response.pivot(index= "Timepoint", columns="Drug", values = "Metastatic Sites")
pivot_merged_two.reset_index(level = None, inplace = True)
# Preview that Reformatting worked
pivot_merged_two.head()
```

```python
reduced_drugs_two= pivot_merged_two[["Timepoint", "Capomulin", "Infubinol", "Ketapril", "Placebo"]]
reduced_drugs_two
```

```python
# Generate the Plot (with Error Bars)
plt.figure(figsize= (10,5))
plt.errorbar(x=reduced_drugs_two['Timepoint'],y=reduced_drugs_two['Capomulin'], yerr=None, linestyle="-", fmt='o', label="Capomulin")
plt.errorbar(x=reduced_drugs_two['Timepoint'],y=reduced_drugs_two['Infubinol'], yerr=None, linestyle="-", fmt='o', label="Infubinol")
plt.errorbar(x=reduced_drugs_two['Timepoint'],y=reduced_drugs_two['Ketapril'], yerr=None, linestyle="-",fmt='o', label="Ketapril")
plt.errorbar(x=reduced_drugs_two['Timepoint'],y=reduced_drugs_two['Placebo'], yerr=None, linestyle="-", fmt='o', label="Placebo")
plt.ylabel("Metastatic Sites")
plt.xlabel("Time(Days)")
plt.title("Metastatic Response to Treatment")
plt.legend()
# Save the Figure

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
merge_fith= merged_data_df.groupby(["Drug", "Timepoint"])["Mouse ID"]
# Convert to DataFrame
survival_rate_count= merge_fith.nunique()
survial_rate_max= 25
survival_rate= (survival_rate_count/survial_rate_max) * 100
survival_rate= pd.DataFrame(survival_rate)
survival_rate.reset_index(inplace=True)
# Preview DataFrame
survival_rate.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivot_merged_three= survival_rate.pivot(index= "Timepoint", columns="Drug", values = "Mouse ID")
pivot_merged_three.reset_index(level = None, inplace = True)
# Preview the Data Frame
pivot_merged_three.head()
```

```python
reduced_drugs_three= pivot_merged_three[["Timepoint", "Capomulin", "Infubinol", "Ketapril", "Placebo"]]
reduced_drugs_three
```

```python
# Generate the Plot (Accounting for percentages)
plt.figure(figsize= (10,5))
plt.errorbar(x=reduced_drugs_three['Timepoint'],y=reduced_drugs_three['Capomulin'], yerr=None, linestyle="-", fmt='o', label="Capomulin")
plt.errorbar(x=reduced_drugs_three['Timepoint'],y=reduced_drugs_three['Infubinol'], yerr=None, linestyle="-", fmt='o', label="Infubinol")
plt.errorbar(x=reduced_drugs_three['Timepoint'],y=reduced_drugs_three['Ketapril'], yerr=None, linestyle="-",fmt='o', label="Ketapril")
plt.errorbar(x=reduced_drugs_three['Timepoint'],y=reduced_drugs_three['Placebo'], yerr=None, linestyle="-", fmt='o', label="Placebo")
plt.ylabel("Survival Rate (%)")
plt.xlabel("Time(Days)")
plt.title("Survival During Treatment")
plt.legend()
# Save the Figure

# Show the Figure
plt.show()
```
